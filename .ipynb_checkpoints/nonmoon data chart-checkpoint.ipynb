```python
import pandas as pd
```

```python
ccr = {"motor_2" : [950,900],
      "motor_5" : [2200,2300],
      "motor_7" : [1850,1950],
      "motor_11" : [1520,1570],
      "motor_12" : [1300,1250],
      "High" : [0.5,5.0]}

ccr_data = pd.DataFrame(ccr)
print(ccr_data)
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams["font.family"] = "Malgun Gothic"
matplotlib.rcParams["font.size"] = 15
matplotlib.rcParams["axes.unicode_minus"] = False
```

```python
fig, axs = plt.subplots(1,2, figsize = (20,10)) # 2 x 2 에 해당하는 plot 들을 생성
fig.suptitle("서보모터의 CCR 변화에 따른 높이 변화")

# 첫번째 그래프
w = 0.15
axs[0].bar(ccr_data.index -2*w, ccr_data["motor_2"], width = w, label = "motor_2")
axs[0].bar(ccr_data.index-w, ccr_data["motor_5"], width = w, label = "motor_5")
axs[0].bar(ccr_data.index, ccr_data["motor_7"], width = w, label = "motor_7")
axs[0].bar(ccr_data.index+w, ccr_data["motor_11"], width = w, label = "motor_11")
axs[0].bar(ccr_data.index+2*w, ccr_data["motor_12"], width = w, label = "motor_12")
axs[0].legend()
axs[0].set_xticks([0,1])

axs[1].bar(ccr_data.index, ccr_data["High"], color = "orange")
axs[1].set_xticks([0,1])
for idx, val in enumerate(ccr_data["High"]):
    axs[1].text(idx, val + 0.08, val, ha = "center")
```
